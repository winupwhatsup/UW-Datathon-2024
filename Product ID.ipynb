```python
#import and read data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/UW 5th datathon/UrbanEdgeApparel.csv")
df
```

```python
#cleaning data
quantity_nan_values = df['Product Quantity'].isna().sum()
selling_nan_values = df['Product Unit Selling Price'].isna().sum()
print(quantity_nan_values)
print(selling_nan_values)

df['Order Date'] = pd.to_datetime(df['Order Date'])
date_sorted = df.sort_values(by='Order Date')
#Order Status should be completed
df[df['Order Status'] == 'Completed']
#Payment should be recieved
df[df['Payment Status'] == 'Received']
```

```python
#Our task is to predict the total amount of sales (quantity/selling price/buying price) for Each product ID
#Pull the data that we need
df = df[['Product ID', 'Order Date', 'Product Quantity','Product Unit Selling Price']]
#selling price
#Let's see selling price for an example product in 10 years span
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Months from Beginning'] = (df['Order Year']- df['Order Year'].min()) * 12 + df['Order Month']
date_sorted = df.sort_values(by='Months from Beginning')
df
```

```python
#test one prod_1000 quantity
prod_1000_data = df[df['Product ID'] == 'Prod_1000']
monthly_quantity = prod_1000_data.groupby('Months from Beginning')['Product Quantity'].sum()
#Plot months from beginning versus quanitiy
plt.figure(figsize=(10, 6))
plt.plot(monthly_quantity.index, monthly_quantity.values, marker='o', linestyle='-')
plt.title(f'Product Quantity Over Months for prod 1000')
plt.grid(True)
plt.show()
#months versus selling price
monthly_avg_price = df.groupby('Months from Beginning')['Product Unit Selling Price'].mean()
#Plot months from beginning versus average selling price
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_price.index, monthly_avg_price.values, marker='o', linestyle='-')
plt.title(f'Selling Price Over Months for prod 1000')
plt.grid(True)
plt.show()
```
